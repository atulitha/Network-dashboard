```python
import pandas as pd
import numpy as np
import xlwt
import requests
import xlrd
import json
```

```python
xl=pd.ExcelFile('dataupdated.xlsx')
df = xl.parse()
df=df.replace(np.nan, "nullvalue")
```

```python
df.columns
```

```python
dict={}
listofState=[]
for i in df.index:
    listofState.append(df['State'][i])
    
listofState=set(listofState)
```

```python
listofState=list(listofState)
```

```python
listofState
```

```python
for i in listofState:
    dp={}
    for j in df.index:
        if(i==df['State'][j]):
            #print(df['New Site Code'][j])
            li=df['New Site Code'][j]+" - "+df['Site Name'][j]
            dp[df['Site Name'][j]]=li
    dict[i]=dp
dict
```

```python
out_file = open("listofsitename.json", "w") 
json.dump(dict, out_file, indent = 4) 
out_file.close() 
```
